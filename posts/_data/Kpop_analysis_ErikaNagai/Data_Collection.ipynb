```python
# Filter only the songs that have more than one artist (= collaboration songs)
# Extract multiple artists

def extract_names(x, key):
    if isinstance(x, str):
        x = eval(x) # convert string to list of dictionaries
    if isinstance(x, list):
        return ', '.join([i.get(key, '') for i in x])
    else:
        return ''

collab_songs['artists_edit'] = collab_songs['artists'].fillna('[]') # replace NaN values with empty list
collab_songs['all_artists'] = collab_songs['artists_edit'].apply(extract_names, key='name') # New column with the name of all artists  
collab_songs['all_artists_id'] = collab_songs['artists'].apply(extract_names, key = 'id') # New column with the id of all artists

collab_songs
```

```python
# Filter out the songs that have only one artist information 
collab_songs = collab_songs[collab_songs['all_artists_id'].str.contains(",")]
collab_songs
```

I need to clean the artists data.

There are some errors in Spotify where the wrong artists with the same or similar name is tagged in the song.
For example, HOYA (capital) sang a song called "Need You Now" with HENRY, however, Hoya (First letter capital and the rest lower case) is wrongly tagged in Spotify.
I manually checked the below rows and replaced the wrong ID and artists name with the correct ones.

```python

# Show duplicated 
artists_df['name_lower'] = artists_df['name'].str.lower().str.strip()
artists_df[artists_df['name_lower'].duplicated(keep=False)].sort_values(['name_lower'])
```

```python
# There are some duplicated artists. In case of duplicates, I want to consider the one with more followers as the main artist
# I need to replace the duplicated artist name with the main artist name, the duplicated id with the mailn artist id in collab_songs and song_detail

# Show the duplicated artists
artists_df[artists_df['name_lower'].duplicated(keep=False)].sort_values(['name_lower'])

# Now make a dataframe of "duplicated_artists" 
duplicated_artists = artists_df[artists_df['name_lower'].duplicated(keep=False)].sort_values(['name_lower'])

# Valid artists are the ones with more followers 
valid_artists = duplicated_artists.loc[duplicated_artists.groupby('name_lower')['followers'].idxmax()]
valid_artists

# Invalid artists are the ones with less followers
invalid_artists = duplicated_artists.loc[duplicated_artists.groupby('name_lower')['followers'].idxmin()]
invalid_artists

# Merge the dataframes 
replaced_artists = pd.merge(
    valid_artists[['name_lower','name', 'id']],
    invalid_artists[['name_lower', 'name', 'id']],
    on = 'name_lower',
    how='outer',
    suffixes=('_valid', '_invalid')
)

replaced_artists['invalid_id'] = replaced_artists['id_invalid'].str.strip()
replaced_artists['valid_id'] = replaced_artists['id_valid'].str.strip()

replaced_artists
```
