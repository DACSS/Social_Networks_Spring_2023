```python
# Gather the "appears_on" songs

import pandas as pd
artist_ids = artists_ids
df1 = pd.DataFrame() # 空のデータフレームを作成

for artist_id in artist_ids:
    # アーティストのアルバム一覧を取得
    results = spotify.artist_albums(artist_id, limit=50, album_type=['album'])
    album_datas = results['items']
    while results['next']:
        results = spotify.next(results)
        album_datas.extend(results['items'])
    
    # 各アルバムからトラック情報を取得し、リストに格納
    track_list = []
    for album_data in album_datas:
        album_id = album_data['id']
        track_data = spotify.album_tracks(album_id)
        for item in track_data['items']:
            track_list.append(item)
    
    # リストからデータフレームを作成
    temp_df1 = pd.DataFrame(track_list)
    
    # アーティスト名の列を追加
    temp_df1['artist_name'] = spotify.artist(artist_id)['name']
    
    # データフレームを連結
    df1 = pd.concat([df1, temp_df1], ignore_index=True)
```

```python
# Clean the data
import pandas as pd
collab_df = pd.read_csv("collab.csv")
# 1. Remove the unnecessary columns 
collab_df = collab_df[['artists', 'id', 'name', 'artist_name', 'artists_str']]
collab_df = collab_df.rename(columns={'name': 'song_name', 'artists': 'api_info', 'artist_name': 'top_kpop_artists', 'artists_str': 'all_artists'})
```

```python
collab_df.dtypes
```

```python
# 2. Remove duplicates
# There are duplicated songs (the same songs but different versions) by the same artists

collab_df['song_10'] = collab_df['song_name'].str[:10]
collab_df = collab_df.drop_duplicates(subset=['song_10', 'all_artists'])

collab_df.sort_values('all_artists')
```

```python
# Now most duplicated songs should have been removed.

# Make some new columns "artist_1" "artist_2"... where the artists in the all_artists are saved separately.


```

```python
# split all_artists into a list of artists

# split all_artists into a list of artists
collab_df['all_artists_list'] = collab_df['all_artists'].str.split(', ')

# remove top_kpop_artists from all_artists_list
collab_df['all_artists_list'] = collab_df.apply(
    lambda row: [artist for artist in row['all_artists_list'] if artist not in row['top_kpop_artists']], axis=1)

# create new columns for each artist in all_artists_list
max_len = max(collab_df['all_artists_list'].apply(len))
for i in range(1, max_len + 1):
    collab_df[f'artist_{i}'] = collab_df['all_artists_list'].apply(lambda x: x[i-1] if len(x) >= i else '')

# drop the temporary all_artists_list column
#collab_df = collab_df.drop(columns='all_artists_list')

collab_df
```

```python
collab_df = collab_df.reindex(columns=['song_name', 'song_10','top_kpop_artists', 'artist_1', 'artist_2', 'artist_3', 'artist_4', 'artist_5', 'all_artists', 'id', 'api_info'])
```
