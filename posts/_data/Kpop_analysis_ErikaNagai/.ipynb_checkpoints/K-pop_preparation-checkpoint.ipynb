```python
import pandas as pd
import numpy as np
```

```python
# 2020 list: According to the data from the Korean Music Content Association (KMCA)

top2020 = """BTS
BLACKPINK
EXO
TWICE
SEVENTEEN
NCT
Red Velvet
GFRIEND
MONSTA X
IU
Stray Kids
Super Junior
MAMAMOO
ITZY
GOT7
(G)I-DLE
ASTRO
OH MY GIRL
SHINee
TXT
IZ*ONE
Apink
EVERGLOW
AB6IX
ATEEZ
N.Flying
BTOB
Park Jin Young
ZICO
DAY6
Hwasa
Changmo
Kang Daniel
The Boyz
Baekhyun
AKMU
SF9
Taeyeon
Taemin
H&D
Kim Ho Joong
BAEKHYUN, SUHO
KANG DANIEL, Simon Dominic, Woo Won-jae
KARD
Jang Beom June
Jessi
Kim Na Young
Kim Jae Hwan
Lee Hi
Lim Young Woong"""

#2015: According to Gaon Music Chart's annual ranking of the top-selling K-pop albums and 
# songs of the year, as well as the monthly rankings for each individual month

top2015 = """Big Bang
EXO
BTS
SNSD (Girls' Generation)
SHINee
Super Junior
Infinite
B2ST (Highlight)
2NE1
GOT7
VIXX
Red Velvet
Apink
f(x)
CNBLUE
AOA
T-ara
4Minute
B1A4
Block B
Teen Top
miss A
SISTAR
NU'EST
WINNER
Lovelyz
MAMAMOO
FTISLAND
EXID
Brown Eyed Girls
Dal Shabet
Kara
2PM
Girls' Day
Secret
Rainbow
BtoB
Wonder Girls
Stellar
Sonamoo
Crayon Pop
G-Friend
ZE:A
History
SPICA
B.A.P
V.O.S
Tahiti
Hello Venus
Royal Pirates"""


# 2010: the annual album sales rankings published by the Hanteo Chart, one of the main music charts in South Korea
top2010 = """Girls' Generation
Super Junior
2NE1
SHINee
Big Bang
Wonder Girls
T-ara
SS501
F.Cuz
BEAST (Highlight)
MBLAQ
After School
CNBLUE
Kara
4Minute
IU
Brown Eyed Girls
Infinite
Secret
2PM
miss A
Rainbow
Davichi
SISTAR
Lee Hyori
Co-Ed School
FTISLAND
GP Basic
ZE:A
D-NA (The Boss)
Teen Top
Choshinsung (Supernova)
S.M. The Ballad
U-KISS
Nine Muses
Seo In Young
TRAX
M to M
JQT
December
2AM
Untouchable
HAM
SeeYa
Lee Seung Gi
Mighty Mouth
Gummy
KARA (Nicole Solo)
Park Bom
CNBLUE (Yonghwa Solo)"""
```

```python
# Create a set of all unique artists mentioned in the three lists
artists_set = set(top2010.split("\n") + top2015.split("\n") + top2020.split("\n"))

# Remove any empty strings or duplicates
artists_set.discard("")
artists_list = list(artists_set)

# Create a dictionary with the artist names and their corresponding years
data = {"Artist": artists_list, "2010": [1 if artist in top2010 else 0 for artist in artists_list],
        "2015": [1 if artist in top2015 else 0 for artist in artists_list],
        "2020": [1 if artist in top2020 else 0 for artist in artists_list]}

# Create the dataframe
df = pd.DataFrame(data)
print(df)
```

## Clean the data

```python
# The below artists' name incllude (xxx), which might cause the mismatch with Kaggle csv

df[df['Artist'].str.contains(r"\((.{2,})\)")]
```

```python


# Choshinsung (Supernova) should be renamed as Supernova
df.loc[112,"Artist"] = "Supernova"

# B2ST is the same as Beast
df.loc[17, "Artist"] = "BEAST"
df.loc[17, "2010"] = 1

# KARA(Nicole Solo), CNBLUE(Yonghwa Solo) and SNSD(Girls' Generation) and BEAST(Highlight) will be removed because those groups already exist in the dataframe
df = df.drop(index=[6,59,83, 115])

# The rest: remove (xx)
df['Artist'] = df['Artist'].str.replace(r"\((.{2,})\)", "")
```
